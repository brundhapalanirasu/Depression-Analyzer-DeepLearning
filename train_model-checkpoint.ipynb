```python
data = pd.read_csv('emotions.csv')
```

```python
import matplotlib.pyplot as plt

label_counts = data['label'].value_counts()

labels = label_counts.index.astype(str)
sizes = label_counts.values

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

plt.axis('equal')
plt.title('Distribution of Labels in Dataset')
plt.show()
```

```python
texts = data['text']        
labels = data['label']  
```

```python
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
import pickle
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
#X_test_vec = vectorizer.transform(X_test)
with open('tfidf_vectorizers.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)
```

```python
model = LogisticRegression(max_iter=1000)  
model.fit(X_train_vec, y_train)

y_pred_lr = model.predict(X_test_vec)
acc_lr = accuracy_score(y_test, y_pred_lr)
print("Classification Report:\n", classification_report(y_test, y_pred_lr))
```

```python
from lightgbm import LGBMClassifier

model = LGBMClassifier()
model.fit(X_train_vec, y_train)

y_pred_lgbm = model.predict(X_test_vec)

acc_lg =  accuracy_score(y_test, y_pred_lgbm)
print("Classification Report:\n", classification_report(y_test, y_pred_lgbm))
```

```python
import pickle

with open('lgbm_model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)

with open('tfidf_vectorizer.pkl', 'rb') as f:
    loaded_vectorizer = pickle.load(f)

new_text = ["im feeling a little like a damaged tree and that my roots are a little out of wack"]
new_text_vec = loaded_vectorizer.transform(new_text)  

prediction = loaded_model.predict(new_text_vec)

prediction_proba = loaded_model.predict_proba(new_text_vec)[0]  # Get first result


label_map = {
    0: 'Anxiety',
    1: 'Bipolar',
    2: 'Depression',
    3: 'Normal',
    4: 'Personality Disorder',
    5: 'Stress',
    6: 'Suicidal'
}

print("Predicted Emotion:", label_map.get(prediction[0], "Unknown"))

print("\nEmotion Probability Breakdown:")
for idx, prob in enumerate(prediction_proba):
    print(f"{label_map[idx]}: {prob * 100:.2f}%")

print(f"\nDepression Probability: {prediction_proba[prediction[0]] * 100:.2f}%")


```
